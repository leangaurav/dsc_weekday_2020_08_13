```python
import pandas as pd
import numpy as np
```

```python
#Read sheet "Australia by Residence" from Australia.xlsx
#Ignore the first 20 rows
data = pd.read_excel("Australia.xlsx",sheet_name='Australia by Residence',skiprows=20)
```

```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
data = data[data.Type != 'Emigrants']
```

```python
data.Type.unique()
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
to_drop = list(data.columns[data.isna().sum()>0.5*data.shape[0]])
data.drop(columns=to_drop,inplace=True)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion)
# Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
print(data.columns)
data.drop(columns=['Type','Coverage','AREA','REG','DEV'],inplace=True,axis=0)
print(data.columns)
```

```python
#Rename columns :OdName -> Country; RegName -> Region
print(data.columns)
data.rename(index=str, columns={"OdName": "Country", "RegName": "Region"},inplace=True)
print(data.columns)
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing 
# "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
print(data.columns.map(type))
data.columns = data.columns.astype(str)
print(data.columns.map(type))

data.apply()
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame
data['Total']= data.sum(axis = 1,numeric_only=True)
data.head(5)
```

```python
# Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)
# find something weird in the dataset ? Delete that row 'world'
df1 = data.sort_values('Total',ascending=False)
#data.head(5)
df1.drop(df1[df1['Country']=='Total'].index,inplace=True)
df1.head(5)
```

```python
# Analyze last few rows also in the sorted result (Don't sort inplace)
data.sort_values('Total',ascending=False).tail(10)
```

```python
##Check if these have all unique values or not:
#countries
#continents (AreaName)
#country continent pairs
print( 'Country : ','There are duplicates' if data['Country'].duplicated().any() else 'No duplicates')
print('AreaName : ', 'There are duplicates' if data['AreaName'].duplicated().any() else 'No duplicates')
print('Country  AreaName combo : ', 'There are duplicates' if data[['AreaName','Country']].duplicated().any() else 'No duplicates')
```

```python
# Find countries for which no of immigrants for all years is 0

data['Country'][data['Total']==0]
```

```python
# Find count of above countries
data['Country'][data['Total']==0].count()
```

```python
# Find number of countries for which no of immigrants is 0 for any year
data.Country[data.eq(0).any(axis = 1)].count()
```

```python
# Drop the row where AreaName is 'World'
data.drop(data[data.AreaName=='World'].index,inplace=True)
data[data.AreaName=='World']
```

```python
# Save the dataset to a new file AustraliaFiltered.xlsx
data.to_excel('AustraliaFiltered.xlsx')
```
