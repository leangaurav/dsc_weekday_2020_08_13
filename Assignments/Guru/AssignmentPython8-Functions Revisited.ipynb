```python
#2 #WAP to use lambda and reduce function to get max number in list
# At first step, first two elements of sequence are picked and the result is obtained.
# Next step is to apply the same function to the previously attained result 
# and the number just succeeding the second element and the result is again stored.
# This process continues till no more elements are left in the container.
# The final returned result is returned and printed on console.

from functools import reduce
s = input("Enter numbers separated by space: ")
l = list(map(int, s.split()))
print(l)
# max_num = reduce(lambda x,y : int(x) if int(x) > int(y) else int(y),l))
max_num = reduce(lambda x,y: x if x > y else y, l)
print("Maximum of {} is {}".format(l,(max_num)))
# 53 48 34 29 35
# 100 23 65 88 32 
```

```python
#7 Assume a list containing heights ft and inches in the form of a list of string 
# example: l=['5ft10in','5ft',..]
# write a function to convert the heights to meter. Use the map function along with your function to 
# convert everything to m. 
# 1ft = 0.3048m
# 1ft = 12in
# 1in = 0.0254m 

# text = '5ft10in'
# ft,inch = tuple(' '.join(re.split('ft|in',text)).split())
# print(ft)
# print(inch)
# print(t)

# text2 = '5ft'
# t2 = ' '.join(re.split('ft|in',text2)).split()
# print(t2)
import re

def con_in_meter(h):
    if 'in' in h:
        ft,inch = tuple(' '.join(re.split('ft|in',h)).split())
#         print(ft, inch)
#         print(type(ft),type(inch))
        total_inch = (int(ft) * 12) + int(inch)
        m = total_inch * 0.0254
    else:
        ft = tuple(' '.join(re.split('ft|in',h)).split())
        ft = ft[0]
        total_inch = (int(ft) * 12)
        m = total_inch * 0.0254
    return(m)

# print(con_in_meter('5ft10in'))
# print(con_in_meter('5ft'))

s = input("Enter heights value in ft and inches separated by space: ")
seq = s.split()
print(list(map(con_in_meter,seq)))
# 5ft10in 5ft 10ft 10ft10in
#Out: [1.778, 1.524, 3.048, 3.302]
```
