```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel("excel_files/AustraliaFiltered.xlsx")
```

```python
# Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants
s=df.sort_values(by="Total",ascending=False).head()
plt.bar(s.Country.values,s.Total.values)
plt.xticks(rotation=90)
plt.show()
```

```python
# Find total immigrants from each category by area name and plot barchart for each area
s=df.groupby("AreaName").Total.sum()
s.plot(kind="bar")
plt.show()

plt.title("MatPlotLib Method")
plt.bar(s.index,s.values)
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
s=df.loc[:,'1980':'2008'].sum()
plt.figure(figsize=(20,7))
s.plot(kind="line")
plt.show()

plt.figure(figsize=(20,7))
plt.title("MatPlotLib Method")
plt.plot(s.index,s.values)
plt.xticks(rotation=90)
plt.show()
```

```python
# Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
s=df.groupby("AreaName").sum().loc['Africa':'Europe','1980':'2008']
plt.figure(figsize=(20,7))
plt.plot(s.columns.values,s.loc["Africa"],label="Africa")
plt.plot(s.columns.values,s.loc["Asia"],label="Asia")
plt.plot(s.columns.values,s.loc["Europe"],label="Europe")
plt.xticks(rotation=90)
plt.legend()
plt.plot()
```

```python
# Plot Total Immigrants based on DevelopmenRegion
s=df.groupby("Region").sum().loc[:,'1980':'2008']
plt.figure(figsize=(20,10))
for i in s.index:
    plt.plot(s.columns.values,s.loc[i].values,label=i)
plt.legend()
plt.show()
```

```python
# Plot immigration trend for India using line plot
s=df.groupby("Country").sum().loc["India",'1980':'2008']
s.plot(kind='line',label="India")
plt.legend()
plt.show()
```

```python
# Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
s=df.sort_values(by="Total",ascending=False).head()
plt.figure(figsize=(20,7))
for i in s.index:
    plt.plot(s.loc[i,"1980":"2008"].index,s.loc[i,"1980":"2008"].values,label=s.iloc[i].Country)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# For the above use area plot with alpha = 0.5
s=df.sort_values(by="Total",ascending=False).head()
s=s.set_index("Country") #Setting country to be the index
s=s.loc[:,'1980':'2008'] #Keeping only the years columns
s=s.T #Transposing - so that x becomes the years and y becomes the values for plotting

s.plot(kind = 'area', stacked = False, alpha = 0.5)
plt.legend(fontsize=5)
plt.show()
```

```python
# Compare Immigration Trend for top-5 countires using stacked area plot
s=df.sort_values(by="Total",ascending=False).head()
s=s.set_index("Country") #Setting country to be the index
s=s.loc[:,'1980':'2008'] #Keeping only the years columns
s=s.T #Transposing - so that x becomes the years and y becomes the values for plotting

s.plot(kind = 'area', stacked = True)
plt.legend(fontsize=5)
plt.show()
```
