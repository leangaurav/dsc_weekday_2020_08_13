```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load data generated previously from 'titanic_filtered.csv'

DF = pd.read_csv('titanic_filtered.csv')
print(DF)
```

```python
#plot histogram of ages of people
plt.figure(figsize=[10,7])
DF.hist(column = 'age')
plt.show()
```

```python
#plot no of people in each pclass on bar chart

c = (DF.pclass.value_counts(sort=False))
print(c)
c.plot(kind='bar')
plt.xlabel('pclass')
plt.show()

plt.bar(c.index,c.values)
plt.show()

c.plot(kind='bar')
plt.xlabel('pclass,pclass')
plt.show()
```

```python
#plot a horizontal bar-chart depicting the ratio of males to females

r = DF.gender.value_counts()
print(r)

r.plot(kind='barh')
plt.show()
```

```python
#plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.

a = DF.groupby(['gender','survived']).survived.count()
print(a)

a.plot(kind='bar')
plt.xlabel('gender,survived')
plt.show()


a.plot(kind='bar',color=['blue','orange','green','red'],width=0.8)
plt.xlabel('gender,survived')
plt.show()
```

```python
#plot a stacked bar chart of number of male-survived,dead and female-survived, dead.

f1 = DF.groupby(['gender','survived'])

f2 = f1.size()

print(f2)
print()
f3 = f2.unstack()

print(f3)

f3.plot(kind='bar',stacked=True)
plt.show()



#or


DF.groupby(['gender','survived']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```
