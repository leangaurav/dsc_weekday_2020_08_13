```python
import pandas as pd
import numpy as np
```

```python
#Read sheet "Australia by Residence" from Australia.xlsx
#Ignore the first 20 rows.
#List all Column and Row indexes(Print their names)

df = pd.read_excel('Australia.xlsx',sheet_name=1,skiprows=20)
print(df)

print(list(df.columns))
print()
print(list(df.index))

```

```python
#Describe the dataset

des = df.describe()
print(des)
```

```python
#Use values and index attributes on both Row and Column Indexes

r = (df.index)
print(r.values)
print()
c = (df.columns)
print(c.values)
```

```python
#Check Shape, Datatypes and other attributes and observe the datatype of each columns' data

print(df.dtypes)
print()
print(df.shape)
print()
print(df.size)
print()
print(df.min())
print()
print(df.max())
print()
print(df.count())
print()
print(df.sum())
```

```python
#Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)

df = df[df['Type']!='Emigrants']
print(df)
```

```python
#Find columns for which more than 50% data is missing and drop them (Column Deletion)

missing_values = df.isnull()
print(missing_values.sum())

#Observation: No column has missing values
```

```python
#Drop Unnecessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"

df = df.drop(axis=1,columns=['Type','Coverage','AREA','REG','DEV'])
print(df)
```

```python
#Rename columns :OdName -> Country; RegName -> Region

df.rename(columns = {'OdName':'Country','RegName':'Region'},inplace=True)
print(df)
```

```python
#Find Total number of immigrants for all years, and create a new column Total in data Frame

Total = df.sum(axis=1)
df['Total'] = Total
print(df)
```
