```python
x = np.arange(10)
print(x)
y = x ** 2
plt.plot(x,y, color='r', marker='*')
plt.title("Plot")
plt.xticks(x)
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()

plt.plot(x,y, color='r', marker='*',alpha=0.2)
plt.title("Plot-2")
plt.xticks(x)
plt.xlabel("Time-2")
plt.ylabel("Price-2")
plt.show()
```

```python
s = pd.Series(np.arange(10)**2)
print(s)
s.plot(kind='line')
plt.xticks(s.index)
plt.show()
```

```python
s = pd.Series(np.arange(10)**2)
print(s)
r = s.plot(kind='bar', width=0.2)
print(r)
plt.xticks(s.index)
plt.show()
```

```python
s = pd.Series(np.arange(10)**2)
print(s)
plt.bar(s.index, s.values)
plt.xticks(s.index)
for i,h in zip(s.index, s.values):
    plt.text(i,h+2 ,str(h))
plt.show()
```

```python
s = pd.Series(np.arange(10)**2)
print(s)
plt.barh(s.index, s.values, tick_label=list("abcdefghij"))# width = 0.5, color='r')
#plt.xticks(s.index)
plt.show()
```

```python
s1 = pd.Series(np.arange(10)**2)
s2 = pd.Series( (np.arange(10)*4)[::-1])
print(s1)
print(s2)
```

```python
plt.bar(s1.index, s1.values)
plt.bar(s2.index, s2.values, bottom=s1.values)
plt.show()
```

```python
r = pd.DataFrame([s1,s2]).T
print(r)

r.plot(kind='bar')
plt.show()

r.plot(kind='bar', stacked=True)
plt.show()
```

```python
r = pd.DataFrame([s1,s2]).T
print(r)
r.index = list("abcdefghij")

r.plot(kind='barh')
plt.show()
```

```python
r = pd.DataFrame([s1,s2]).T
r['A'] = pd.Series(list("abcdefghij"))
print(r)
t = r.set_index('A')
print(t)
t.plot(kind='bar')
```
